```python
import cv2
import os

cap = cv2.VideoCapture("./TestVid/5020.mp4")

output = "./TestVid/"

frame_count = 0
while frame_count < 60:
    ret, frame = cap.read()
    if not ret:
        break  # video ended before 60 frames

    # save frame as PNG
    frame_filename = os.path.join(output, f"00{frame_count:04d}.png")
    cv2.imwrite(frame_filename, frame)

    frame_count += 1

cap.release()
print(f"Saved {frame_count} frames to {output}")
```
